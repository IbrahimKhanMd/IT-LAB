```python
def calculate_cv_sv(EV, PV, AC):
    CV = EV - AC
    SV = EV - PV
    return CV, SV

EV, PV, AC = 5000, 6000, 5500
CV, SV = calculate_cv_sv(EV, PV, AC)
print(f"Cost Variance (CV) = {CV}")
print(f"Schedule Variance (SV) = {SV}")
```

```python
def calculate_cpi_spi(EV, PV, AC):
    CPI = EV / AC
    SPI = EV / PV
    budget_status = "Under Budget" if CPI > 1 else "Over Budget"
    schedule_status = "Ahead of Schedule" if SPI > 1 else "Behind Schedule"
    return CPI, SPI, budget_status, schedule_status

EV, PV, AC = 8000, 7500, 8500
CPI, SPI, budget_status, schedule_status = calculate_cpi_spi(EV, PV, AC)
print(f"CPI = {CPI:.2f} ({budget_status})")
print(f"SPI = {SPI:.2f} ({schedule_status})")
```

```python
def calculate_eac(BAC, EV, AC, PV):
    CPI = EV / AC
    SPI = EV / PV

    EAC1 = BAC / CPI

   
    EAC2 = AC + (BAC - EV)

   
    EAC3 = AC + (BAC - EV) / (CPI * SPI)


    recommended_EAC = EAC1
    recommendation = "Based on current CPI trend"

    return EAC1, EAC2, EAC3, recommended_EAC, recommendation

BAC, EV, AC, PV = 100000, 40000, 45000, 50000
EAC1, EAC2, EAC3, rec_EAC, rec_msg = calculate_eac(BAC, EV, AC, PV)

print(f"EAC Method 1 (BAC/CPI) = {EAC1:.2f}")
print(f"EAC Method 2 (AC + BAC - EV) = {EAC2:.2f}")
print(f"EAC Method 3 (AC + (BAC - EV)/(CPI*SPI)) = {EAC3:.2f}")
print(f"Recommended EAC = {rec_EAC:.2f} ({rec_msg})")
```

Write a Python program that acts as an interactive Earned Value Management (EVM) calculator. The program should:

Take user inputs for Planned Value (PV), Earned Value (EV), and Actual Cost (AC).

Compute the following EVM metrics:

Cost Variance (CV)

Schedule Variance (SV)

Cost Performance Index (CPI)

Schedule Performance Index (SPI)

Estimate at Completion (EAC) using the formula: EAC = BAC / CPI, where BAC = PV * 2 (simplified assumption).

Interpret the results to indicate whether the project is under/over budget and ahead/behind schedule.

Display the outputs in a user-friendly format.

Input:

Enter Planned Value (PV): 76000  
Enter Earned Value (EV): 74000  
Enter Actual Cost (AC): 69000  

Expected Output:

=== EVM Analysis Results ===  
- Planned Value (PV): $76,000  
- Earned Value (EV): $74,000  
- Actual Cost (AC): $69,000  

--- Variance Analysis ---  
- Cost Variance (CV): $5,000 (Under Budget)  
- Schedule Variance (SV): $-2,000 (Behind Schedule)  

--- Performance Indices ---  
- CPI: 1.07 (Cost Efficient)  
- SPI: 0.97 (Schedule Lagging)  

--- Forecast ---  
- Estimate at Completion (EAC): $93,243  
- Projected Savings/Overrun: $-7,243 (Overrun)  

## Final code - EVM simulator 

```python
def evm_calculator():
    
    PV = float(input("Enter Planned Value (PV): "))
    EV = float(input("Enter Earned Value (EV): "))
    AC = float(input("Enter Actual Cost (AC): "))

    CV = EV - AC
    SV = EV - PV
    CPI = EV / AC
    SPI = EV / PV
    BAC = PV * 2 
    EAC = BAC / CPI

    budget_status = "Under Budget" if CV > 0 else "Over Budget"
    schedule_status = "Ahead of Schedule" if SV > 0 else "Behind Schedule"


    print("\n=== EVM Analysis Results ===")
    print(f"- Planned Value (PV): ${PV:,.2f}")
    print(f"- Earned Value (EV): ${EV:,.2f}")
    print(f"- Actual Cost (AC): ${AC:,.2f}\n")

    print("--- Variance Analysis ---")
    print(f"- Cost Variance (CV): ${CV:,.2f} ({budget_status})")
    print(f"- Schedule Variance (SV): ${SV:,.2f} ({schedule_status})\n")

    print("--- Performance Indices ---")
    print(f"- CPI: {CPI:.2f} ({'Cost Efficient' if CPI > 1 else 'Cost Inefficient'})")
    print(f"- SPI: {SPI:.2f} ({'Schedule Efficient' if SPI > 1 else 'Schedule Lagging'})\n")

    print("--- Forecast ---")
    print(f"- Estimate at Completion (EAC): ${EAC:,.2f}")
    print(f"- Projected Savings/Overrun: ${BAC - EAC:,.2f} ({'Savings' if (BAC - EAC) > 0 else 'Overrun'})")

evm_calculator()
```
